<li> Compute AUC Score, you need to compute different thresholds and for each threshold compute tpr,fpr and then use               numpy.trapz(tpr_array, fpr_array) <a href='https://stackoverflow.com/q/53603376/4084039'>https://stackoverflow.com/q/53603376/4084039</a>, <a href='https://stackoverflow.com/a/39678975/4084039'>https://stackoverflow.com/a/39678975/4084039</a> Note: it should be numpy.trapz(tpr_array, fpr_array) not numpy.trapz(fpr_array, tpr_array)</li>

```python
data = pd.read_csv("/content/drive/MyDrive/Assignment 5/Copy of 5_a.csv")
data1 = data.copy()
data1.columns
data1.y.value_counts()
yp=[]
for i in data1.proba:
  if i > 0.5:
    yp.append(1)
  else:
    yp.append(0)
data1.insert(2,'yp',yp)
data1.drop(['proba'],axis=1,inplace=True)
TP = len((data1.query('y==1 & yp==1')).axes[0])
TN = len((data1.query('y==0 & yp==0')).axes[0])
FP = len((data1.query('y==0 & yp==1')).axes[0])
FN = len((data1.query('y==1 & yp==0')).axes[0])
#Compute Confu matrix
confu_matrx = np.array([[TP,FP],[FN,TN]])
print(f"Confusion Matrix: \n {confu_matrx} \n")
#Compute F1 Score
precision = TP/(TP+FP)
recall = TP/(FN+TP)
f1_score = 2 * ((precision*recall)/(precision+recall))
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1_Score: {f1_score} \n")
#Compute Accuracy
accuracy = (TP+TN)/(TP+FP+TN+FN)
print(f"Accuracy: {accuracy}")
```

```python
#Compute AUC Score
data = pd.read_csv("/content/drive/MyDrive/Assignment 5/Copy of 5_a.csv")
data5= data.sort_values("proba",ascending=False).copy()
print(data5.columns)
TPR_arr =[]
FPR_arr = []
for i in data5.proba:
  yp1 = []
  for j in data5.proba:
    if j > i:
      yp1.append(1)
    else:
      yp1.append(0)
  data5.insert(2,'yp1',yp1)
  TP = len((data5.query('y==1 & yp1==1')).axes[0])
  TN = len((data5.query('y==0 & yp1==0')).axes[0])
  FP = len((data5.query('y==0 & yp1==1')).axes[0])
  FN = len((data5.query('y==1 & yp1==0')).axes[0])
  try:
    TPR = TP/(TP+FP)
  except ZeroDivisionError:
    TPR = 0
  FPR = FP/(TN+FP)
  TPR_arr.append(TPR)
  FPR_arr.append(FPR)
  data5.drop(['yp1'],axis=1,inplace=True)
print(f"TPR : {TPR_arr}")
print(f"FPR: {FPR_arr}")
AUC_score = np.trapz(TPR_arr,FPR_arr)
print(f"AUC_score: {AUC_score} \n")
```

<ol>

<li> Compute MAPE: https://www.youtube.com/watch?v=ly6ztgIkUxk</li>
<li> Compute R^2 error: https://en.wikipedia.org/wiki/Coefficient_of_determination#Definitions </li>
</ol>

```python
data = pd.read_csv("/content/drive/MyDrive/Assignment 5/Copy of 5_d.csv")
data3 = data.copy()
data3.head()
#Compute Mean Square Error
ei = (data3.y - data3.pred)
msr = sum(np.square(ei))/len(data3)
print(f"Mean Squared Error : {msr}")
```

```python
#Compute Mean Absolute Percentage Error
mape = np.sum(np.abs(data3.y-data3.pred))/np.sum(data3.y)
print(f"MAPE: {mape}")
```

```python
#Compute R^2 Error
y_mean = np.mean(data3.y)
SS_res = np.sum(np.square(ei))
SS_total = np.sum(np.square(data3.y-y_mean))
R_2 = 1- (SS_res/SS_total)
print(f"R^2 Error: {R_2}")
```
